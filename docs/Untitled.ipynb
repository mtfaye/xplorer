```python
def sampling(processed_df, ss=20000):
    return processed_df.sample(ss)
    

def simple_corr(sampling):
    return sampling.corr(method='pearson')


def _encoder(sampling):
    return pd.get_dummies(sampling, columns=None, drop_first=True)
    

def select_target(sampling, col):
    target = sampling.col
    return target
```

```python
def crr(_encoder, col, k=10):
    c = _encoder.corr().nlargest(k, col)[col].index
    cm = _encoder[c].corr()
    return cm  
```

```python
sm = sampling(df, ss=20000)
_encoder(sm)
```

```python
col = "Amount"
enc = _encoder(sm)

crr(enc, col, k=30)
```
